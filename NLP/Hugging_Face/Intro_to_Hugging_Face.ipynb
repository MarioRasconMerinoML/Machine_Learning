Libraries
Learn about Hugging Face’s open source libraries, including Transformers and Diffusers.

Key ideas:

Hugging Face’s open source work includes libraries to simplify the use of pre-trained ML models
Two such libraries are Transformers and Diffusers. Transformer models are great at text-based tasks, while diffuser models are generally used for image-based tasks.
The Hugging Face libraries allow developers to quickly import and work with a large variety of pre-trained transformer and diffuser models, like GPT2 (an early version of the ChatGPT model) or Stable Diffusion.
For example, here’s all 5 lines of code needed to import Stable Diffusion (a standard image generation library) and use it to generate an image of a squirrel in the style of Picasso:

from diffusers import DiffusionPipeline
import torch

pipeline = DiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16)
pipeline.to("cuda")
pipeline("An image of a squirrel in Picasso style").images[0]

# Spaces

Learn about Hugging Face Hub’s Spaces feature for easily building and hosting ML demo apps.

Key ideas:

Spaces is a feature on Hugging Face Hub for building and hosting ML demo apps.
- Using Spaces allows an ML developer to quickly create a demo app with a user interface for interacting with a model.
- Building a demo on Spaces requires only a few lines of code using Python packages like Gradio and Streamlit.
- Note that while basic Spaces are free, more advanced Spaces (like those requiring GPUs) may require a paid upgrade.
